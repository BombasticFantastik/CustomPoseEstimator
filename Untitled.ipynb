```python
колво слов длины 1 - 5
колво слов длины 2 - 9
колво слов длины 3 - 29

выходит формула bn=(bn-1)+(4bn-2)

переписываем это на Python:
last_last_b=5
last_b=9
for n in range (2,13):
    correct_b=last_b+4*last_last_b
    print(correct_b)
    last_last_b=last_b
    last_b=correct_b
    
```
